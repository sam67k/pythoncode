```python

print("******************inputs for matrix**********************")
row = int(input("Enter Number of Rows : "))
col = int(input("Enter Number of Columns : "))
matrix1 = [[0]*col for i in range(row)]

print()

print("\n*****************matrix-1 created************************\n")

for r in range(row):
    print("| ",end='')
    for c in range(col):
        print(matrix1[r][c]," ",end='')
    print("|")

print("\n*****************inputs for matrix-1 elements*********************\n")

for r in range(row):
    for c in range(col):
        print("\nEnter Element [",r,"] [",c,"] for Matrix-1")
        matrix1[r][c] = int(input()) 

print("\n***********************matrix-1 after giving elements values****************************\n")

for r in range(row):
    print("| ",end='')
    for c in range(col):
        print(matrix1[r][c]," ",end='')
    print("|")

matrix2 = [[0]*col for i in range(row)]

print("\n\n***********************matrix-2 created**********************************\n")

for r in range(row):
    print("| ",end='')
    for c in range(col):
        print(matrix2[r][c]," ",end='')
    print("|")
       
print("\n***************************inputs for matrix-2 elements********************************\n")
    
for r in range(row):
    for c in range(col):
        print("\nEnter Element [",r,"] [",c,"] for Matrix-2")
        matrix2[r][c] = int(input()) 

print("\n***************************matrix-2 after giving elements values********************************\n")

for r in range(row):
    print("| ",end='')
    for c in range(col):
        print(matrix2[r][c]," ",end='')
    print("|")

matrix3 = [[0]*col for i in range(row)]
    
for r in range(row):
    for c in range(col):
        matrix3[r][c] = matrix1[r][c] + matrix2[r][c]

print("\n\n**********************Answer (After Addition)*************************************\n")

for r in range(row):
    print("| ",end='')
    for c in range(col):
        print(matrix3[r][c]," ",end='')
    print("|")
```

```python
print("******************inputs for matrix-1**********************")
row1 = int(input("Enter Number of Rows : "))
col1 = int(input("Enter Number of Columns : "))
matrix1 = [[0]*col1 for i in range(row1)]

print()

print("\n*****************matrix-1 created************************\n")

for r in range(row1):
    print("| ",end='')
    for c in range(col1):
        print(matrix1[r][c]," ",end='')
    print("|")

print("\n*****************inputs for matrix-1 elements*********************\n")

for r in range(row1):
    for c in range(col1):
        print("\nEnter Element [",r,"] [",c,"] for Matrix-1")
        matrix1[r][c] = int(input()) 

print("\n***********************matrix-1 after giving elements values****************************\n")

for r in range(row1):
    print("| ",end='')
    for c in range(col1):
        print(matrix1[r][c]," ",end='')
    print("|")

print("\n******************inputs for matrix-2**********************")
row2 = int(input("Enter Number of Rows : "))
col2 = int(input("Enter Number of Columns : "))
matrix2 = [[0]*col2 for i in range(row2)]

print("\n\n***********************matrix-2 created**********************************\n")

for r in range(row2):
    print("| ",end='')
    for c in range(col2):
        print(matrix2[r][c]," ",end='')
    print("|")
       
print("\n***************************inputs for matrix-2 elements********************************\n")
    
for r in range(row2):
    for c in range(col2):
        print("\nEnter Element [",r,"] [",c,"] for Matrix-2")
        matrix2[r][c] = int(input()) 

print("\n***************************matrix-2 after giving elements values********************************\n")

for r in range(row2):
    print("| ",end='')
    for c in range(col2):
        print(matrix2[r][c]," ",end='')
    print("|")

if(row1 == col2):
    print("\nMatrix are able to be Multiplied\n")  

matrix3 = [[0]*col2 for i in range(row1)]   
    
for i in range(row1):
    for j in range(col2):
        for k in range(col2):
            matrix3[i][j] += matrix1[i][k] * matrix2[k][j]
            
for r in range(row1):
    print("| ",end='')
    for c in range(col2):
        print(matrix3[r][c]," ",end='')
    print("|")
```
